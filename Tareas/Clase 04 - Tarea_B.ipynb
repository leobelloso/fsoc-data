## Punto 2

#### ¿Qué hipotesis puede plantear de este set de datos respecto a la variable stroke?

- Los fumadores y ex-fumadores son más propensos a padecer ACV
- Las personas con mayor índice de masa corporal y glucosa en sangre son más propensos a padecer un ACV
- Las personas con mayor índice de glucosa en sangre y que fueron o son fumadores son más propensas a padecer un ACV
- Las personas hipertensas que ya hayan padecido algun tipo de enfermedad coronaria son más propensas a padecer un ACV

## Punto 3

#### ¿Qué gráficos utilizaría y por qué?

- Para la primer hipótesis, utilizaría un gráfico de barras puesto que necesito contar la cantidad de frecuencias en cada caso y compararla con las frecuencias de "ACV" hay por categoría. Podría representarse de la siguiente forma:

```python
sns.countplot(data=df,x='smoking_status', hue='stroke')
plt.title("Casos ACV según Carácter de Fumador")
plt.xlabel("Carácter de Fumador")
plt.ylabel("Frecuencia")
plt.show()
```

- Para la segunda hipótesis, utilizaría un histograma para observar si un alto IMC y un alto promedio de glucosa en sangre tienen más chances de provocar un ACV. **Utilicé un histograma porque estoy trabajando con variables de valores continuos.** Podría representarse de la siguiente forma:

```python
sns.histplot(data=df,x='avg_glucose_level', y='bmi', hue='stroke')


plt.title("Posibilidades de ACV según Media de Glucosa en Sangre e IMC")
plt.xlabel("Media de Glucosa en Sangre")
plt.ylabel("Índice de Masa Corporal")
plt.grid()
plt.show()
```

- La tercera hipótesis puede representarse con un gráfico de barras. Si bien, los premedios de glucosa son valores continuos, la condición de fumador es una variable cualitativa nominal, por eso la utilización de este tipo de gráfico.

```python
sns.barplot(data=df,x='smoking_status', y='avg_glucose_level', hue='stroke')


plt.title("Posibilidades de ACV según Condición de Fumador y Media de Glucosa en Sangre")
plt.xlabel("Condición de Fumador")
plt.ylabel("Media de Glucosa en Sangre")
plt.grid()
plt.show()
```

- Al tratarse de dos variables discretas, puede usarse un gráfico de líneas para establecer si hay una relación entre ambas.

```python
sns.lineplot(data=df,x='hypertension', y='heart_disease', hue='stroke')


plt.title("Posibilidades de ACV según Condición de Fumador y Media de Glucosa en Sangre")
plt.xlabel("Hipertensión")
plt.ylabel("Enfermedad Coronaria")
plt.grid()
plt.show()
```
